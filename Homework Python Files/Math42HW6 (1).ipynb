```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
G = np.array([[1.170773, 0], [0, 1.0720542]])
```

```python
def population_model(initial):
    return G @ initial
```

```python
pop_model_cal = [np.array((12988), dtype=float)]
pop_model_rest = [np.array((152082), dtype=float)]
initial = np.array([[12988], [152082]])
for i in range(12):
    pop_model_cal.append(population_model(initial)[0])
    pop_model_rest.append(population_model(initial)[1])
    initial = population_model(initial)
```

```python
years = list(range(1955, 2020, 5))
```

```python
actual_pop_cal = (1 / 1000) * np.array([15717204, 19953134, 23667902, 29760021, 33871648, 37253956, 39538223], dtype = float)
years_by_10 = np.array([1960, 1970, 1980, 1990, 2000, 2010, 2020], dtype = float)
```

```python
plt.plot(years, pop_model_cal)
plt.plot(years_by_10, actual_pop_cal)
plt.title("Population Model for California")
plt.xlabel("Year")
plt.ylabel("Population (in thousands)")
plt.legend(labels = ["model", "actual"])
plt.show()
```

```python
actual_pop_rest = (1 / 1000) * np.array([179323175, 203211926, 226545805, 248709873, 281421906, 308745538, 331449281]) - actual_pop_cal
```

```python
plt.plot(years, pop_model_rest)
plt.plot(years_by_10, actual_pop_rest)
plt.title("Population Model for Rest of U.S.")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(labels = ["model", "actual"])
plt.show()
```
